```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
grid = np.linspace(-3,3,500)

def funcao1(x):
    return 16*x*(math.pi - x)/(5*math.pi**2-4*x*(math.pi-x))

def funcao2(x):
    return np.sin(x)

def funcao3(x):
    return x - x**3/6 + x**5/120

y1=funcao1(grid)
y2=funcao2(grid)
y3=funcao3(grid)

plt.xlabel('x')

plt.plot(grid,y1, label='16*x*(pi - x)/(5*pi**2-4*x*(pi-x))')
plt.plot(grid,y2, label='sin(x)')
plt.plot(grid,y3, label='x - x^3/6 + x*^5/120')
plt.legend()
plt.tight_layout()

plt.savefig('senos')
```
